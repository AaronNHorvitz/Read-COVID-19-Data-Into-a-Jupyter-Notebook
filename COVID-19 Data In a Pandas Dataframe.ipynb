# dependencies

```python
import pandas as pd
```

# settings

```python
pd.options.display.max_rows = 500        #Increases the number of rows displayed without a break in the middle.
pd.options.display.max_columns = 100     #Increases the number of columns displayed without a break in the middle.
pd.set_option('max_colwidth',100)        #Increases the width of the columns to display 100 characters at the same time.
```

# data location

This is the online location of the raw data posted on Github by Johns Hopkins University:

Confirmed Cases:  
https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv

Deaths:  
https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv

Recovered: 
https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv

```python
# assign the url locations

url_confirmed = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
url_deaths    = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv'
url_recovered = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'
```

# read the data into a dataframe

```python
# read the datainto a dataframe

confirmed = pd.read_html(url_confirmed)[0]
deaths    = pd.read_html(url_deaths)[0]
recovered = pd.read_html(url_recovered)[0]

# drop the first column. 

confirmed = confirmed.drop(columns = 'Unnamed: 0')
deaths    = deaths.drop(columns = 'Unnamed: 0')
recovered = recovered.drop(columns = 'Unnamed: 0')
```

# display the data

```python
print('\nConfirmed Cases:')
display(confirmed.head())

print('\nDeaths:')
display(deaths.head())

print('\nRecovered:')
display(recovered.head())
```
